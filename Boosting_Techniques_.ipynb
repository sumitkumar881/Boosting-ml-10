```python
# Q1.What is Boosting in Machine Learning?
Ans:-Boosting is a machine learning ensemble meta-algorithm for reducing bias and variance in supervised learning, and a family of machine learning algorithms that convert weak learners into strong ones. Boosting is based on the idea that a collection of weak learners can be combined to create a strong learner. A weak learner is a model that is only slightly better than random guessing, such as a decision stump (a decision tree with only one level). A strong learner is a model that is significantly better than random guessing.

Boosting works by iteratively training weak learners on weighted versions of the data. In each iteration, the weights of the data points that were misclassified by the previous weak learner are increased. This forces the next weak learner to focus on the data points that were most difficult to classify. After several iterations, the weak learners are combined to create a strong learner.

There are several different boosting algorithms, such as AdaBoost, Gradient Boosting, and XGBoost. These algorithms differ in the way they weight the data points and combine the weak learners. However, they all share the same basic idea of iteratively training weak learners on weighted versions of the data.

Boosting is a powerful technique that can be used to improve the performance of a wide variety of machine learning algorithms. It is particularly useful for problems where the data is highly imbalanced or where the decision boundary is complex.

# Q2.How does Boosting differ from Bagging?
Ans:-In Bagging, multiple models are trained independently on different random subsets of the training data (created by sampling with replacement). The final prediction is typically an average (for regression) or a majority vote (for classification) of the individual model predictions. The main goal of Bagging is to reduce variance and prevent overfitting.

models are trained sequentially. Each subsequent model is trained to correct the errors made by the previous models. Data points that were misclassified by the previous models are given higher weights in the next iteration, forcing the new model to focus on these difficult examples. The final prediction is a weighted combination of the individual model predictions. The main goal of Boosting is to reduce bias and convert weak learners into strong learners.

# Q3.What is the key idea behind AdaBoost?
Ans:-The key idea behind AdaBoost (Adaptive Boosting) is to sequentially train weak learners (typically decision stumps) on weighted versions of the training data, where the weights are adjusted in each iteration to give more importance to misclassified data points. The final strong learner is a weighted sum of the weak learners, with the weights of the weak learners being proportional to their accuracy. It focuses on turning a collection of simple, "weak" classifiers into a single, highly accurate "strong" classifier by giving more focus to the examples that are harder to classify.

# Q4.Explain the working of AdaBoost with an example?
Ans:-Imagine you have a dataset of points, some are circles (positive class) and some are crosses (negative class). Your goal is to build a classifier that can accurately separate circles from crosses.

**Step 1: Initialize weights**
Initially, every data point is given the same weight.

**Step 2: Train a weak learner (Decision Stump)**
Train a simple decision stump (a decision tree with only one split) on the weighted data. This stump tries to find a simple rule (like "if x > 0.5") to separate the classes. It will likely make some mistakes.

**Step 3: Calculate the error of the weak learner**
Calculate the error of the weak learner based on the weighted data. Data points that were misclassified contribute more to the error due to their higher weights.

**Step 4: Calculate the "influence" of the weak learner**
Based on the error, calculate a weight (alpha) for this weak learner. The more accurate the weak learner is (lower error), the higher its weight will be. This weight represents how much this weak learner will influence the final decision.

**Step 5: Update data point weights**
Increase the weights of the data points that were misclassified by the current weak learner. This makes those misclassified points more important for the next weak learner. Decrease the weights of correctly classified points.

**Step 6: Repeat**
Repeat steps 2-5 for a fixed number of iterations (or until some stopping criterion is met). In each iteration, a new weak learner is trained on the adjusted weights, and its weight is calculated.

**Step 7: Combine weak learners**
The final strong classifier is a weighted sum of all the trained weak learners. The weight of each weak learner is the alpha calculated in Step 4. For a new data point, each weak learner makes a prediction, and these predictions are combined based on the weak learners' weights to arrive at the final classification.

# Q5.What is Gradient Boosting, and how is it different from AdaBoost?
Ans:-# What is Gradient Boosting?
Gradient Boosting is a machine learning technique that builds an ensemble of weak prediction models, typically decision trees. It works by sequentially adding predictors (models) to the ensemble. Each new predictor is trained on the "residuals" or "errors" of the previous ensemble. Instead of re-weighting data points like AdaBoost, Gradient Boosting fits the new model to the negative gradient of the loss function. The goal is to minimize the errors and improve the accuracy of the overall ensemble. Think of it as fixing the mistakes of the previous models in a clever, mathematical way.

# How is it different from AdaBoost?

Here are the key differences between Gradient Boosting and AdaBoost:

*   **Focus of Correction:**
    *   **AdaBoost:** Corrects misclassified *data points* by increasing their weights for the next weak learner.
    *   **Gradient Boosting:** Corrects the *residuals* (the difference between the predicted value and the actual value) by fitting the new model to these errors. It focuses on reducing the loss function.

*   **Weak Learner Fitting:**
    *   **AdaBoost:** Fits the weak learner to the original data distribution, but with adjusted weights for each data point.
    *   **Gradient Boosting:** Fits the weak learner to the residual errors of the previous ensemble.

*   **Loss Function:**
    *   **AdaBoost:** Primarily uses an exponential loss function.
    *   **Gradient Boosting:** Can use a variety of differentiable loss functions (e.g., mean squared error for regression, log loss for classification). This flexibility makes it applicable to a wider range of problems.

*   **Parallelization:**
    *   **AdaBoost:** Training is inherently sequential because the weights for the next iteration depend on the errors of the current iteration.
    *   **Gradient Boosting:** Similar to AdaBoost, the core boosting process is sequential, but some implementations (like XGBoost, LightGBM) introduce techniques for parallelization within the tree building process.

*   **Sensitivity to Outliers:**
    *   **AdaBoost:** Can be sensitive to noisy data and outliers because it focuses on misclassified points, and outliers are often misclassified.
    *   **Gradient Boosting:** Can be less sensitive to outliers depending on the chosen loss function.

# Q6.What is the loss function in Gradient Boosting?
Ans:-The loss function in Gradient Boosting is a differentiable function that quantifies the error between the predicted values and the actual values. The specific loss function used depends on the type of problem being solved:

*   **For Regression:**
    *   **Mean Squared Error (MSE):** $L(y, F(x)) = (y - F(x))^2$. This is the most common loss function for regression and penalizes larger errors more heavily.
    *   **Mean Absolute Error (MAE):** $L(y, F(x)) = |y - F(x)|$. This is less sensitive to outliers than MSE.
    *   **Huber Loss:** A combination of MSE and MAE, less sensitive to outliers than MSE but still smooth.
    *   **Quantile Loss:** Used for quantile regression.

*   **For Classification:**
    *   **Log Loss (Binary Cross-Entropy):** $L(y, F(x)) = - (y \log(p) + (1-y) \log(1-p))$, where $p$ is the predicted probability. This is the standard loss for binary classification.
    *   **Multiclass Log Loss (Categorical Cross-Entropy):** Used for multiclass classification.
    *   **Exponential Loss:** While AdaBoost primarily uses this, it can also be used in Gradient Boosting.

Gradient Boosting aims to minimize this loss function by iteratively adding weak learners that fit the negative gradient of the loss function with respect to the current predictions. This effectively means each new tree is trained to predict the residual errors of the previous ensemble, with the goal of reducing the overall error as measured by the chosen loss function.

# Q7.How does XGBoost improve over traditional Gradient Boosting?
Ans:-XGBoost (Extreme Gradient Boosting) is an optimized, distributed, and efficient gradient boosting library. It was developed with the goal of pushing the limits of computation power for boosted tree algorithms and has become widely popular for its speed and performance. While traditional Gradient Boosting is a powerful algorithm, XGBoost introduces several key improvements:

1.  **Regularization:** XGBoost includes L1 (Lasso) and L2 (Ridge) regularization terms in its objective function. This helps to prevent overfitting, which can be a significant issue in traditional Gradient Boosting, especially with deep trees. The regularization term penalizes the complexity of the model, encouraging simpler trees.

2.  **Parallel Processing:** XGBoost supports parallel processing. While the core boosting process is sequential, XGBoost parallelizes the *tree construction* process. When building each tree, it can parallelize the finding of the best split points for features. This significantly speeds up training, especially on large datasets.

3.  **Handling Missing Values:** XGBoost has a built-in mechanism to handle missing values. It can learn the best direction for missing values (to go to the left or right child node) based on training data, rather than requiring explicit imputation.

4.  **Tree Pruning:** XGBoost implements a "depth-first" approach to tree building and uses post-pruning. It grows trees up to a maximum depth and then prunes back branches that do not improve the loss function by a certain threshold (controlled by the `gamma` parameter). Traditional Gradient Boosting often uses a "level-wise" approach without explicit pruning, which can be less efficient.

5.  **Hardware Optimization:** XGBoost is designed with hardware optimization in mind. It utilizes techniques like cache-aware access and off-core computation to improve efficiency and speed.

6.  **Sparsity Awareness:** XGBoost can handle sparse data efficiently. It has an optimized algorithm for sparse feature matrices, which is particularly useful in datasets with many zero values.

7.  **Weighted Quantile Sketch:** For finding optimal split points for continuous features, XGBoost uses a distributed weighted quantile sketch algorithm. This is more efficient than enumerating all possible split points, especially for large datasets.

8.  **Flexibility:** While not unique to XGBoost compared to all Gradient Boosting implementations, XGBoost offers flexibility in choosing different objective functions and evaluation metrics.

# Q8.What is the difference between XGBoost and CatBoost?
Ans:- XGBoost                                       CatBoost

| Requires pre-processing (e.g., one-hot)     | Native, built-in handling (Ordered Target Encoding)
| Can be prone to target leakage              | Designed to prevent target leakage
| Can sometimes occur                         | Mitigated by Ordered Boosting
| Traditional Gradient Descent                | Ordered Boosting
| Asymmetric trees                            | Symmetric (Oblivious) trees
| Often require tuning                        | Well-tuned, good out-of-the-box performance
| Generally very fast                         | Can be slightly slower
| Fast                                        | Often very fast (especially with symmetric trees)
| Built-in L1 and L2                          | Built-in (different approach)
| Built-in handling                           | Built-in handling

# Q9.What are some real-world applications of Boosting techniques?
Ans:-**Spam Detection:** Boosting is widely used to build highly accurate spam filters by combining the predictions of multiple weaker classifiers that look for different spam characteristics (e.g., specific keywords, sender information, email structure).
*   **Fraud Detection:** In finance and insurance, boosting algorithms are employed to identify fraudulent transactions or claims by learning complex patterns from large datasets.
*   **Image and Object Recognition:** Boosting has been successfully applied in computer vision for tasks like face detection (famously in the Viola-Jones algorithm which used AdaBoost) and recognizing other objects in images.
*   **Medical Diagnosis:** Boosting can be used to improve the accuracy of medical diagnostic models by combining information from various symptoms, test results, and patient history.
*   **Recommendation Systems:** Boosting can enhance recommendation engines by predicting user preferences based on historical data and combining the outputs of different recommendation strategies.
*   **Search Ranking:** Search engines often utilize boosting techniques to rank search results based on their relevance to a user's query, combining signals from various factors like page content, links, and user behavior.
*   **Credit Scoring:** Financial institutions use boosting models to assess the creditworthiness of individuals by combining information from various financial indicators.
*   **Predictive Maintenance:** In industrial settings, boosting can be used to predict when equipment is likely to fail, allowing for proactive maintenance and reducing downtime.
*   **Natural Language Processing (NLP):** Boosting can be applied to various NLP tasks like sentiment analysis, text classification, and named entity recognition.
*   **Customer Churn Prediction:** Businesses use boosting to predict which customers are likely to stop using their service, allowing them to implement retention strategies.

# Q10.How does regularization help in XGBoost?
AnS:-Regularization in XGBoost helps to prevent overfitting by penalizing complex models. XGBoost includes both L1 (Lasso) and L2 (Ridge) regularization terms in its objective function.

Here's how it helps:

1.  **Reduces Model Complexity:** Regularization adds a penalty term to the loss function that is being minimized. This penalty increases with the complexity of the model.
    *   **L1 Regularization (Lasso):** Adds a penalty proportional to the absolute value of the weights of the tree's leaves. This can lead to sparsity, meaning it can force some leaf weights to become exactly zero, effectively simplifying the tree structure.
    *   **L2 Regularization (Ridge):** Adds a penalty proportional to the square of the weights of the tree's leaves. This discourages large weights, making the model less sensitive to individual data points and leading to smoother predictions.

2.  **Controls Tree Depth and Number of Leaves:** The regularization parameters (`reg_alpha` for L1 and `reg_lambda` for L2) influence the tree splitting process. A higher regularization penalty makes it harder for the algorithm to split a node, effectively controlling the depth of the trees and the number of leaves. Simpler trees are less likely to capture noise in the training data.

3.  **Improves Generalization:** By limiting the complexity of the individual trees and the overall model, regularization helps the model generalize better to unseen data. Without regularization, XGBoost models can become very complex and memorize the training data, leading to poor performance on new data.

In essence, regularization in XGBoost acts as a constraint that prevents the model from becoming too specialized to the training data. It strikes a balance between fitting the training data well and keeping the model simple enough to perform well on future data. This is crucial for building robust and reliable predictive models.

# Q11.What are some hyperparameters to tune in Gradient Boosting models?
Ans:-Here are some important hyperparameters to tune in Gradient Boosting models, particularly focusing on algorithms like XGBoost, LightGBM, and CatBoost, which are common implementations:

**Parameters Controlling the Overall Boosting Process:**

1.  **`n_estimators` (or `num_boost_round` in XGBoost):** The number of boosting stages or the number of trees to build. Increasing this can improve performance but also increases the risk of overfitting and computational cost. Often tuned using early stopping based on a validation set.

2.  **`learning_rate` (or `eta` in XGBoost):** Controls the step size shrinkage. After each tree is added, the model's predictions are updated with the contribution of the new tree, scaled by the learning rate. A smaller learning rate requires more `n_estimators` but makes the model more robust to overfitting and can lead to better final performance. This is a very critical parameter to tune.

3.  **`subsample`:** The fraction of samples used to train each tree. Using a value less than 1.0 introduces randomness and helps prevent overfitting. Typical values are between 0.6 and 1.0.

4.  **`colsample_bytree` (or `feature_fraction` in LightGBM):** The fraction of features used when building each tree. Similar to subsample, this adds randomness and helps prevent overfitting. You can also have variations like `colsample_bylevel` and `colsample_bynode` for more fine-grained control.

**Parameters Controlling Individual Trees (Weak Learners):**

5.  **`max_depth`:** The maximum depth of each tree. Deeper trees can capture more complex interactions but are more prone to overfitting. Shallow trees (like decision stumps, max_depth=1) are very weak learners.

6.  **`min_child_weight` (or `min_data_in_leaf` in LightGBM):** The minimum sum of weights of all observations required in a child node to consider a split. This helps prevent the growth of trees in regions of sparse data and smooths the learned model. A higher value makes the model more conservative.

7.  **`gamma` (or `min_split_loss` in XGBoost):** The minimum loss reduction required to make a further partition on a leaf node of the tree. This effectively controls tree pruning. A higher value makes the algorithm more conservative and less likely to split.

8.  **`reg_alpha` (L1 regularization) and `reg_lambda` (L2 regularization) in XGBoost:** Regularization terms on the weights of the leaves. These help prevent overfitting by penalizing the complexity of the model. Higher values lead to more regularization.

**Parameters Specific to CatBoost:**

9.  **`l2_leaf_reg`:** Coefficient of L2 regularization term on weights. Similar to `reg_lambda` in XGBoost.

10. **`border_count`:** The number of splits for continuous features. Controls the quantization of continuous features.

11. **`one_hot_max_size`:** Features with a number of unique values less than or equal to this parameter are one-hot encoded.

12. **`cat_features`:** Parameter to explicitly specify which features are categorical. CatBoost handles categorical features natively using ordered target encoding.

**Other Parameters:**

13. **`objective` (or `loss`):** The loss function to be minimized. The choice depends on the problem (e.g., 'reg:squarederror' for regression, 'binary:logistic' for binary classification, 'multi:softprob' for multiclass classification in XGBoost).

14. **`eval_metric`:** The metric used to evaluate the performance during training (e.g., 'rmse', 'mae', 'logloss', 'auc'). This is often used with early stopping.

# Q12.What is the concept of Feature Importance in Boosting?
Ans:-The concept of **Feature Importance** in Boosting refers to how much each feature (or input variable) contributes to the final prediction of the boosted model. Since boosting algorithms build an ensemble of weak learners (often decision trees), Feature Importance is typically calculated by aggregating the importance of each feature across all the trees in the ensemble.

Here's the breakdown:

1.  **Importance within a single tree:** In a decision tree, a feature's importance is usually determined by how much it reduces the impurity (or loss) at the nodes where it's used for splitting. The more a feature is used for splitting and the greater the reduction in impurity it provides, the more important that feature is considered within that tree. Common metrics include:
    *   **Gain/Impurity Reduction:** The total reduction in the loss function achieved by splits based on a particular feature across all the trees. This is the most common measure.
    *   **Split Count:** The number of times a feature is used to split a node across all the trees.
    *   **Weight:** The number of times a feature appears in the splitting conditions of the trees.

2.  **Aggregating across trees:** The importance scores from individual trees are then averaged or summed up across the entire ensemble to get a global feature importance score for the boosted model. Features that consistently contribute significantly to reducing the error in multiple trees will have higher overall importance scores.

**Why is Feature Importance useful in Boosting?**

*   **Interpretability:** It helps understand which features are most influential in the model's predictions, providing insights into the underlying relationships in the data.
*   **Feature Selection:** Features with low importance can potentially be removed without significantly impacting model performance, leading to simpler and potentially faster models, and mitigating the risk of overfitting.
*   **Domain Understanding:** Identifying important features can reveal valuable insights about the domain or problem you are trying to solve.
*   **Model Debugging:** Unexpectedly low or high importance for certain features can indicate issues with the data, feature engineering, or the model itself.

# Q13. Why is CatBoost efficient for categorical data?
Ans:-CatBoost is particularly efficient and effective for handling categorical features primarily due to two key innovations:

1.  **Ordered Target Encoding (or Ordered Boosting):** Traditional methods for handling categorical features in gradient boosting (like one-hot encoding or standard target encoding) can suffer from target leakage, especially on small datasets. Target leakage occurs when information from the target variable "leaks" into the features used for training, leading to overly optimistic performance on the training set and poor generalization to new data.

    CatBoost addresses this with a sophisticated ordering principle. When calculating the target-mean for a categorical feature, it uses a dynamic, history-based approach. Instead of calculating the mean of the target for all instances of a category using the *entire* dataset (which introduces leakage), it calculates the target mean for a category using only the data points that came *before* the current data point in a randomly permuted order.

    Specifically, for a categorical feature and a specific data point, the encoded value is calculated based on the average of the target values of the training examples that have the same category value and appeared *earlier* in a random permutation of the training data, combined with a prior. This prevents the model from seeing the target value of the current instance (or instances appearing later in the order) when encoding its categorical features, thus significantly reducing target leakage. This ordered approach is applied both during the calculation of the splits (using the encoded values) and in the boosting process itself (Ordered Boosting).

2.  **Native Handling without Explicit Pre-processing:** CatBoost's algorithm is designed to work directly with categorical features, eliminating the need for manual pre-processing steps like one-hot encoding or pre-calculating target means. You simply specify which columns are categorical, and CatBoost handles the encoding internally using its ordered target encoding scheme during training. This simplifies the data preparation pipeline and reduces the risk of errors introduced by manual encoding methods.
```

```python
# Q14.Train an AdaBoost Classifier on a sample dataset and print model accuracy.
!pip install scikit-learn

from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification

# Generate a sample dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, random_state=42)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the AdaBoost Classifier
adaboost = AdaBoostClassifier(n_estimators=50, random_state=42)
adaboost.fit(X_train, y_train)

# Make predictions on the test set
y_pred = adaboost.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"AdaBoost Classifier Accuracy: {accuracy}")
```

```python
# Q15.Train an AdaBoost Regressor and evaluate performance using Mean Absolute Error (MAE)
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.datasets import make_regression

# Generate a sample regression dataset
X_reg, y_reg = make_regression(n_samples=1000, n_features=20, n_informative=10, random_state=42)

# Split data into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)

# Initialize and train the AdaBoost Regressor
adaboost_reg = AdaBoostRegressor(n_estimators=50, random_state=42)
adaboost_reg.fit(X_train_reg, y_train_reg)

# Make predictions on the test set for regression
y_pred_reg = adaboost_reg.predict(X_test_reg)

# Calculate and print the Mean Absolute Error
mae = mean_absolute_error(y_test_reg, y_pred_reg)
print(f"AdaBoost Regressor Mean Absolute Error (MAE): {mae}")
```

```python
# Q27.Train an XGBoost Regressor and tune the learning rate using GridSearchCV
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generate a sample regression dataset (if not already available from Q20)
# from sklearn.datasets import make_regression
# from sklearn.model_selection import train_test_split
X_reg, y_reg = make_regression(n_samples=1000, n_features=20, n_informative=10, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)


# Initialize the XGBoost Regressor
# We'll tune the learning_rate, so set n_estimators to a reasonably large number
# GridSearchCV will cross-validate, so we don't need a separate test set for tuning splits
xgb_reg_gs = xgb.XGBRegressor(objective='reg:squarederror', # Corrected objective function name
                            n_estimators=200, # Use a fixed reasonable number for tuning learning_rate
                            max_depth=3,      # Fix other common parameters or include them in the grid
                            random_state=42)

# Define the parameter grid for learning_rate
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3]
}

# Initialize GridSearchCV
# Using X_train_reg and y_train_reg for cross-validation
# cv=5 means 5-fold cross-validation
# scoring='neg_mean_squared_error' because GridSearchCV maximizes the score,
# so we use the negative of MSE to find the minimum MSE.
grid_search = GridSearchCV(estimator=xgb_reg_gs,
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error', # Use negative MSE
                           cv=5,
                           verbose=1,
                           n_jobs=-1) # Use all available cores

# Fit GridSearchCV to the training data
grid_search.fit(X_train_reg, y_train_reg)

# Print the best parameters and the best score
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation negative MSE: {grid_search.best_score_:.4f}")
# Convert negative MSE to positive MSE for easier interpretation
print(f"Best cross-validation MSE: {-grid_search.best_score_:.4f}")

# You can access the best model trained by GridSearchCV
best_xgb_reg = grid_search.best_estimator_

# Evaluate the best model on the unseen test set
y_pred_best_reg = best_xgb_reg.predict(X_test_reg)
final_mse = mean_squared_error(y_test_reg, y_pred_best_reg)
print(f"Mean Squared Error on test set with best parameters: {final_mse:.4f}")
```

```python
# Q28.Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting
import matplotlib.pyplot as plt
!pip install catboost imbalanced-learn

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, f1_score
from collections import Counter
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# 1. Generate an imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5,
                           weights=[0.95], # 95% majority class, 5% minority class
                           flip_y=0.05, # Some noise
                           random_state=42)

print(f"Original dataset shape: {Counter(y)}")

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # Stratify to maintain imbalance ratio

print(f"Training set shape: {Counter(y_train)}")
print(f"Test set shape: {Counter(y_test)}")

# 2. Train CatBoost without class weighting
print("\n--- Training CatBoost without Class Weighting ---")
catboost_no_weight = CatBoostClassifier(iterations=100,
                                        learning_rate=0.1,
                                        depth=3,
                                        loss_function='Logloss',
                                        eval_metric='F1', # Use F1 since it's better for imbalanced data
                                        random_state=42,
                                        verbose=0)

catboost_no_weight.fit(X_train, y_train,
                       eval_set=(X_test, y_test),
                       early_stopping_rounds=10,
                       verbose_eval=False)

y_pred_no_weight = catboost_no_weight.predict(X_test)

print("\nClassification Report (No Class Weighting):")
print(classification_report(y_test, y_pred_no_weight))
f1_no_weight = f1_score(y_test, y_pred_no_weight)
print(f"F1 Score (No Class Weighting): {f1_no_weight:.4f}")

# 3. Train CatBoost with class weighting
print("\n--- Training CatBoost with Class Weighting ---")

# Calculate class weights
# We can use `compute_class_weight` from sklearn.utils
# This will give higher weights to the minority class
classes = np.unique(y_train)
class_weights = compute_class_weight('balanced', classes=classes, y=y_train)
class_weights_dict = dict(zip(classes, class_weights))

print(f"Calculated Class Weights: {class_weights_dict}")

catboost_with_weight = CatBoostClassifier(iterations=100,
                                          learning_rate=0.1,
                                          depth=3,
                                          loss_function='Logloss',
                                          eval_metric='F1',
                                          random_state=42,
                                          class_weights=class_weights_dict, # Apply class weights
                                          verbose=0)

catboost_with_weight.fit(X_train, y_train,
                         eval_set=(X_test, y_test),
                         early_stopping_rounds=10,
                         verbose_eval=False)

y_pred_with_weight = catboost_with_weight.predict(X_test)

print("\nClassification Report (With Class Weighting):")
print(classification_report(y_test, y_pred_with_weight))
f1_with_weight = f1_score(y_test, y_pred_with_weight)
print(f"F1 Score (With Class Weighting): {f1_with_weight:.4f}")

# 4. Compare performance
print("\n--- Performance Comparison ---")
print(f"F1 Score (No Class Weighting):   {f1_no_weight:.4f}")
print(f"F1 Score (With Class Weighting): {f1_with_weight:.4f}")

# Visualize the confusion matrices
cm_no_weight = confusion_matrix(y_test, y_pred_no_weight)
cm_with_weight = confusion_matrix(y_test, y_pred_with_weight)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(cm_no_weight, annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')
axes[0].set_title('Confusion Matrix (No Class Weighting)')

sns.heatmap(cm_with_weight, annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[1],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')
axes[1].set_title('Confusion Matrix (With Class Weighting)')

plt.tight_layout()
plt.show()
```
